```python
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

from preprocessing import build_preprocessor

df = pd.read_csv("../data/student-mat.csv", sep=";")
df.head()
```

```python

X_base = df.drop("G3", axis=1)
y = df["G3"]

X_train, X_test, y_train, y_test = train_test_split(
    X_base, y, test_size=0.2, random_state=42
)

preprocessor = build_preprocessor(X_train)

pipeline = Pipeline([
    ("preprocessing", preprocessor),
    ("model", LinearRegression())
])

pipeline.fit(X_train, y_train)

y_pred_base = pipeline.predict(X_test)

score_base = r2_score(y_test, y_pred_base)

print("Base model R2:", score_base)
```

```python
df["study_per_absence"] = df["studytime"] / (df["absences"] + 1)
```

Giải thích: vì học nhiều nhưng nghỉ nhiều chưa chắc đã tốt, học đều và ít nghỉ thường hiểu quả cao hơn. kết hợp 2 yếu tố học và nghỉ để tạo feature mới tốt hơn

```python
df["failure_impact"] = df["failures"] * df["absences"]
```

Giải thích: rớt và nghỉ đều ảnh hưởng tiêu cực đến điểm cuối kỳ. kết hợp 2 yếu tố bằng cách nhân sẽ khuếch đại khả năng điểm thấp của học sinh

```python
X_new = df.drop("G3", axis=1)

X_train, X_test, y_train, y_test = train_test_split(
    X_new, y, test_size=0.2, random_state=42
)

preprocessor = build_preprocessor(X_train)

pipeline = Pipeline([
    ("preprocessing", preprocessor),
    ("model", LinearRegression())
])

pipeline.fit(X_train, y_train)

y_pred_new = pipeline.predict(X_test)

score_new = r2_score(y_test, y_pred_new)

print("New model R2:", score_new)
```
