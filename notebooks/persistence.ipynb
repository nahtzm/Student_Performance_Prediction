```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
import datetime
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import learning_curve

from preprocessing import load_and_split
```

```python
_, x_test, _, y_test = load_and_split("../data/student-mat.csv")

model_path = '../models/best_model.pkl'
best_model = joblib.load(model_path)

print(f"Đã load mô hình từ: {model_path}")

y_pred = best_model.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
```

```python
final_package = {
    "pipeline": best_model,
    "metadata": {
        "date_created": datetime.datetime.now().strftime("%Y-%m-%d %H:%M"),
        "metrics": {"R2": r2, "MAE": mae, "RMSE": rmse},
        "description": "Model dự đoán điểm G3 dựa trên thông tin học tập.."
    }
}
save_path = '../models/model_package.pkl'
joblib.dump(final_package, save_path)
print(f"Đã lưu sản phẩm cuối cùng tại: {save_path}")
```

```python
demo_data = joblib.load('../models/model_package.joblib')
demo_pipeline = demo_data["pipeline"]
demo_meta = demo_data["metadata"]

print(f"Thông tin Model: {demo_meta['description']}")

# test
sample_student = x_test.iloc[[5]]
predicted_score = demo_pipeline.predict(sample_student)

print(f"Dự đoán điểm G3: {predicted_score[0]:.2f}")
print(f"Điểm thực tế G3: {y_test.iloc[0]}")
print("\nPipeline đã được lưu và tải lên thành công.")
```
