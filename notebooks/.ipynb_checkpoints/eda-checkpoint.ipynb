```python
# Biểu đồ 8: Phát hiện giá trị ngoại lai trong số buổi nghỉ học
plt.figure(figsize=(6,4))

sns.boxplot(x=df["absences"])

plt.title("Phát hiện giá trị ngoại lai trong số buổi nghỉ học")

plt.xlabel("Số buổi nghỉ học")

plt.show()
```

Nhận xét: Có một số học sinh có số buổi nghỉ học rất cao so với phần lớn dữ liệu, những giá trị này có thể được xem là outliers và cần xem xét trong quá trình tiền xử lý dữ liệu

```python

num_cols = ['age', 'failures', 'absences', 'traveltime']

for col in num_cols:
    plt.figure(figsize=(7,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
def count_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = data[(data[col] < lower) | (data[col] > upper)]
    
    return len(outliers), lower, upper

for col in num_cols:
    count, lower, upper = count_outliers_iqr(df, col)
    print(f"{col}: {count} outliers (Lower={lower:.2f}, Upper={upper:.2f})")
```

Nhận xét: Có 1 học sinh 22 tuổi có thể do học muộn hoặc lưu ban, không ảnh hưởng. Giá trị 4 của traveltime bị đánh dấu là ngoại lệ theo IQR nhưng vẫn nằm trong phạm vi hợp lệ của dữ liệu. Có vài học sinh nghỉ học nhiều hơn rất nhiều so với các bạn khác có thể nghỉ học lun. Các giá trị nghỉ học cực đoan (>20) sẽ được xử lý bằng capping để giảm ảnh hưởng của ngoại lệ nhưng vẫn giữ nguyên số lượng bản ghi. Biến failuers có nhiều outliers thực chất là do dataset bị lệch  phân phối mạnh về 0 nên vẫn hợp lệ.
