```python
import wandb
wandb.login()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

from preprocessing import load_and_split, build_preprocessor


X_train, X_test, y_train, y_test = load_and_split("../data/student-mat.csv")


preprocessor = build_preprocessor()


configs = [
    {"n_estimators": 100, "max_depth": None},
    {"n_estimators": 100, "max_depth": 5},
    {"n_estimators": 200, "max_depth": None},
    {"n_estimators": 200, "max_depth": 5},
    {"n_estimators": 300, "max_depth": 10},
]

for cfg in configs:

    wandb.init(
        project="student-performance",
        config={
        "model": "RandomForestRegressor",
        "n_estimators": cfg["n_estimators"],
        "max_depth": cfg["max_depth"],
        "random_state": 42
    }

    )


    pipeline = Pipeline([
        ("preprocessing", preprocessor),
        ("model", RandomForestRegressor(
            n_estimators=cfg["n_estimators"],
            max_depth=cfg["max_depth"],
            random_state=42
        ))
    ])


    pipeline.fit(X_train, y_train)


    y_pred = pipeline.predict(X_test)


    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)


    wandb.log({
        "MAE": mae,
        "RMSE": rmse,
        "R2": r2
    })


    plt.figure(figsize=(6, 6))
    plt.scatter(y_test, y_pred)
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
    plt.title("Actual vs Predicted")

    wandb.log({
        "Prediction Chart": wandb.Image(plt)
    })

    plt.close()


    wandb.finish()
```

Đã thực hiện 5 thí nghiệm để so sánh hiệu suất các mô hình.

Experiment 3: RMSE = 1.89811, MAE = 1.4367, R² = 0.8243. Kết quả tốt nhất.
Experiment 4: RMSE = 2.02677, MAE = 1.52923, R² = 0.79967. Kết quả thấp nhất.
Experiment 5: RMSE = 1.89621, MAE = 1.44346, R² = 0.82465. Kết quả rất tốt và ổn định.
